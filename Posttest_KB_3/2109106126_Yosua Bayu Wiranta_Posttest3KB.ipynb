```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

Employee = pd.read_csv('Total.csv') #pendeklarasian untuk membaca data
Employee.head()
```

## 1. Grafik Batang


Plot batang memvisualisasikan distribusi 10 jabatan karyawan teratas di San Francisco berdasarkan kumpulan data yang disediakan. Sumbu x mewakili jabatan, dan sumbu y mewakili jumlah karyawan yang memegang setiap jabatan.

```python
#Mengelompokkan data berdasarkan tahun dan menghitung jumlah data per tahun
data_by_year = Employee['Year'].value_counts().sort_index()

# Membuat grafik batang
plt.figure(figsize=(10, 6))
sns.barplot(x=data_by_year.index, y=data_by_year.values, palette='viridis')
plt.title('Jumlah Data Gaji per Tahun di San Francisco')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Data')

# Menampilkan grafik batang
plt.show()
```

## 2. Pie Chart

Diagram lingkaran memvisualisasikan distribusi gaji karyawan di San Francisco berdasarkan kumpulan data yang disediakan. Setiap bagian mewakili rentang gaji, dan ukuran setiap bagian sesuai dengan persentase karyawan dalam rentang gaji tersebut relatif terhadap jumlah total karyawan dalam kumpulan data.

```python
# Selecting top 10 job titles for illustration
top_job_titles = Employee['JobTitle'].value_counts().head(10)

# Plotting the pie chart
plt.figure(figsize=(10, 10))
plt.pie(top_job_titles, labels=top_job_titles.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Top 10 Employee Job Titles in San Francisco')
plt.show()
```

## 3. Grafik Line Plot


Plot garis memvisualisasikan gaji rata-rata karyawan dari waktu ke waktu berdasarkan kumpulan data yang disediakan. Sumbu x mewakili tahun, dan sumbu y mewakili gaji rata-rata. Garis yang menghubungkan penanda menggambarkan tren atau perubahan gaji rata-rata selama bertahun-tahun.

```python
lineplot_data = Employee.groupby('Year')['TotalPay'].mean().reset_index()

# Plotting the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='TotalPay', data=lineplot_data, marker='o', color='skyblue', linestyle='-', linewidth=2)
plt.title('Average Employee Salary Over Time')
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.grid(True)
plt.show()
```

## 4. Heatmap

Peta panas memvisualisasikan matriks korelasi fitur numerik dalam kumpulan data gaji karyawan San Francisco. Setiap sel di peta panas mewakili koefisien korelasi antara dua variabel. Warna menunjukkan kekuatan dan arah korelasi: warna hangat (misalnya merah) menunjukkan korelasi positif, warna dingin (misalnya biru) menunjukkan korelasi negatif, dan putih menunjukkan tidak ada korelasi.

```python
numeric_columns = Employee.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap - San Francisco Employee Salaries')
plt.xlabel('Numerical Features')
plt.ylabel('Numerical Features')
plt.show()
```

## 5. Histogram

Histogram memberikan representasi visual distribusi gaji karyawan di San Francisco berdasarkan dataset yang disediakan. Sumbu x mewakili rentang gaji, dan sumbu y mewakili frekuensi atau jumlah karyawan dalam setiap rentang gaji.

```python
plt.figure(figsize=(10, 6))
sns.histplot(Employee['TotalPay'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Employee Salaries in San Francisco')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```
